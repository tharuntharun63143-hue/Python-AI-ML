```python
## step-1: import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
## step-2: creating a small dataset
data = {'Experience':[1,2,3,4,5,6,7],
        'Salary': [35000,40000,45000,50000,60000,65000,70000]}
df = pd.DataFrame(data)
df
## step-3: visualize data
plt.scatter(df['Experience'], df['Salary'], color='blue')
plt.title('Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
```

```python
##Step-4: train linear regression Model
x = df[['Experience']]
y = df['Salary']
model = LinearRegression()
model.fit(x,y)
##Step-5: view model parameters
print("slops(m):",model.coef_)
print("intercept(c):",model.intercept_)
```

```python
##step-6 predicte the new values
predicted = model.predict(x)
print(predicted)
```

```python
##plot regression 
plt.scatter(x,y, color='green')
plt.plot(x,predicted, color='blue',linewidth=2)
plt.title('Linear Regression Best fit line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
```

```python
##step-8: Predict for new data
new_exp = np.array([[8]])
predicted_salary = model.predict(new_exp)
print(f"Predicted Salary for 8 years of experience: ₹{predicted_salary[0]:.2f}")
```

```python
##House Price Predict Dataset
##Step-1 : Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = {'sqrt.ft':[100,200,300,800,1000],
       'price':[10000,20000,30000,400000,50000]}
df = pd.DataFrame(data)
df
## step-3: visualize data
plt.scatter(df['sqrt.ft'], df['price'], color='blue')
plt.title("House sqrt.ft price")
plt.xlabel('sqrt.ft of House')
plt.ylabel('House price')
plt.show
```

```python
##step-4 train linear regression model
x = df[['sqrt.ft']]
y = df['price']
model = LinearRegression()
model.fit(x,y)
##Step-5: view model parameters
print("slops(m):",model.coef_)
print("intercept(c):",model.intercept_)
```

```python
##step-5: predict of house sqrt.ft
predicted = model.predict(x)
print(predicted)
```

```python
##step-6: linear Regression
plt.scatter(x,y, color='blue')
plt.plot(x,predicted, color='green',linewidth=2)
plt.title('Linear Regression Best fit line')
plt.xlabel('Sqrt.ft of house')
plt.ylabel('price')
plt.show
```

```python
##step-7: Predict of the house price
house_price = np.array([[2000]])
predicted_houseprice = model.predict(house_price)
print(f"Predicted House Price for  2000sqrt.ft: ₹{predicted_houseprice[0]:.2f}")
```
