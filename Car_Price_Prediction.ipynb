```python
##Step-1:Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
## step-2: load dataset 
data = pd.read_csv("carprice.csv")
print(data.head())
```

```python
## step-3: selecting features
x = data[['stroke']]# independent variable
y = data[['price']] # dependent variable 
## step-4: splitting data
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.5, random_state=23)
```

```python
##step-5: train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
##step-6
print("Slop(m):",model.coef_)
print("intercept(c):",model.intercept_)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_test,y_test, color='yellow', label='Actual price')
plt.plot(x_test, y_pred, color='red',linewidth=1, label='Predicted line')
plt.title('Linear regression fit- Car prices')
plt.xlabel('stroke(Car)')
plt.ylabel('Price')
plt.legend()
plt.show
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score

mae = mean_absolute_error(y_test, y_pred)
mse =mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE:",mae)
print("MSE",mse)
print("r2 Score",r2)
```
