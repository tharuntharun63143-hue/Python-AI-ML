```python
##step-1: import libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
## step-2: load dataset 
data = pd.read_csv("final_dataset.csv")
print(data.head())
```

```python

## step-3: selecting features
x = data[['size']]# independent variable
y = data[['price']] # dependent variable 
## step-4: splitting data
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
##step-5: train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("slope(m):", model.coef_)
print("intercept(c):", model.intercept_)
```

```python
##Step-6:prediction and testing the model
y_pred = model.predict(x_test)
```

```python
##Step-7:Visulation the Model
plt.scatter(x_test,y_test, color='red', label='Actual price')
plt.plot(x_test, y_pred, color='orange',linewidth=2, label='Predicted line')
plt.title('Linear regression fit- house prices')
plt.xlabel('Size(sqft)')
plt.ylabel('Price')
plt.legend()
plt.show
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score

mae = mean_absolute_error(y_test, y_pred)
mse =mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)

print("MAE:",mae)
print("MSE",mse)
print("r2 Score",r2)
```
